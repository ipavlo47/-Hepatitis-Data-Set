```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import requests
import os
import seaborn as sns
```

```python
file_path = "hepatitis.data"
df = pd.DataFrame()
df = pd.read_csv(file_path, delimiter=",", header=None)
df.rename(columns={0: 'Class', 1: 'AGE', 2: 'SEX', 3: 'STEROID', 4: 'ANTIVIRALS', 5: 'FATIGUE', 6: 'MALAISE',
                  7: 'ANOREXIA', 8: 'LIVER BIG', 9: 'LIVER FIRM', 10: 'SPLEEN PALPABLE', 11: 'SPIDERS', 
                  12: 'ASCITES', 13: 'VARICES', 14: 'BILIRUBIN', 15: 'ALK PHOSPHATE', 16: 'SGOT',
                  17: 'ALBUMIN', 18: 'PROTIME', 19: 'HISTOLOGY',}, inplace=True)
df
```

```python
df.mean()
```

```python
df = df.fillna(df.mean())
df
```

```python
df = df.round({'STEROID':0, 'ANTIVIRALS':0, 'FATIGUE':0, 'MALAISE':0, 'ANOREXIA':0, 
          'LIVER BIG':0, 'LIVER FIRM':0, 'SPLEEN PALPABLE':0, 'SPIDERS':0, 
          'ASCITES':0, 'VARICES':0, 'BILIRUBIN':5, 'ALK PHOSPHATE':0, 'SGOT':0,
                  'ALBUMIN':1, 'PROTIME':0, 'HISTOLOGY':0,}) 
df
```

```python
for k in ['BILIRUBIN', 'ALK PHOSPHATE', 'SGOT','ALBUMIN', 'PROTIME']:
    max = float(df[k].max())
    min = float(df[k].min())
    for i in range(0, 155):
        '''print(k, i , df[k][i], min, max)
        print((float(df[k][i]) - min)/(max - min))'''
        df[k][i] = (float(df[k][i]) - min)/(max - min)
        
df    
```

```python
hist_plot = df['BILIRUBIN'].hist(bins=40)
hist_plot.set_title('Bilirubin level in time of hepatits')
hist_plot.set_xlabel('Bilirubin level in the blood')
hist_plot.set_ylabel('Frequency')
```

```python
hist_plot = df['ALK PHOSPHATE'].hist(bins=40)
hist_plot.set_title('Alk phosphate level in time of hepatits')
hist_plot.set_xlabel('Alk phosphate level in the blood')
hist_plot.set_ylabel('Frequency')
```

```python
hist_plot = df['SGOT'].hist(bins=40)
hist_plot.set_title('SGOT level in time of hepatits')
hist_plot.set_xlabel('SGOT level in the blood')
hist_plot.set_ylabel('Frequency')
```

```python
hist_plot = df['ALBUMIN'].hist(bins=40)
hist_plot.set_title('Albumin level in time of hepatits')
hist_plot.set_xlabel('Albumin level in the blood')
hist_plot.set_ylabel('Frequency')
```

```python
hist_plot = df['PROTIME'].hist(bins=40)
hist_plot.set_title('Protime level in time of hepatits')
hist_plot.set_xlabel('Protime level in the blood')
hist_plot.set_ylabel('Frequency')
```

```python
df.plot(x='AGE', y='SGOT', kind='line')
```

```python
cols = ['BILIRUBIN', 'ALK PHOSPHATE', 'SGOT','ALBUMIN', 'PROTIME']
pp = sns.pairplot(df[cols], height=1.8, aspect=1.8,
                  plot_kws=dict(edgecolor="k", linewidth=0.5),
                  diag_kind="kde", diag_kws=dict(shade=True))

fig = pp.fig 
fig.subplots_adjust(top=0.93, wspace=0.3)
t = fig.suptitle('Dependence', fontsize=14)
```

```python
df.corr(method='pearson')
```

```python
f, ax = plt.subplots(figsize=(20, 14))
corr = df.corr(method='pearson')
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=0.025)
f.subplots_adjust(top=0.93)
t= f.suptitle('Pearson Correlation Heatmap', fontsize=14)
```

```python
df.corr(method='spearman')
```

```python
f, ax = plt.subplots(figsize=(20, 14))
corr = df.corr(method='spearman')
hm = sns.heatmap(round(corr,2), annot=True, ax=ax, cmap="coolwarm",fmt='.2f',
                 linewidths=0.025)
f.subplots_adjust(top=0.93)
t= f.suptitle('Spearman Correlation Heatmap', fontsize=14)
```
